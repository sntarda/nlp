```python
# Read the dataset
def load_data(file_path):
    """
    Load the dataset from a file, supporting TXT and SQL files.
    Parameters:
    - file_path: str, the path to the dataset file.
    Returns:
    - text_data: str, the content of the dataset file.
    """
    text_data = ''
    if file_path.endswith('.sql'):
        # Assuming the SQL file contains insert commands with Quranic text
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                # Extracting text within single quotes in SQL insert statements
                matches = re.findall(r"'(.*?)'", line)
                if matches:
                    text_data += ' '.join(matches) + ' '
    else:
        with open(file_path, 'r', encoding='utf-8') as file:
            text_data = file.read()
    return text_data
```

# **Cleaning and Normalization**

```python
import pandas as pd
import re
import os
import spacy
from spacy.lang.ar.examples import sentences
from spacy.lang.ar import Arabic

# Intialize SpaCy tokenizer
nlp = spacy.blank('ar')

# Function to tokenize text using the spaCy Arabic pipeline
def tokenize(text):
    # Process the text with the NLP pipeline
    doc = nlp(text)
    # Extract tokens as strings
    tokens = [token.text for token in doc]
    return tokens

# Load the dataset
# Assuming the text of interest is in the first column, adjust as necessary
clean_quran_df = QuranText.copy()

# Function to clean and normalize Arabic text (as you've defined)

def clean_normalize_quran_text(text):
    """
    Clean and normalize the Arabic text data.
    Parameters:
    - text: str, the Arabic text to be cleaned and normalized.
    Returns:
    - cleaned_text: str, the cleaned and normalized text.
    """
    # Remove HTML tags
    cleaned_text = re.sub('<[^<]+?>', '', text)

    # Normalize Arabic characters (e.g., Alif forms, Hamza)
    cleaned_text = re.sub('[إأٱآا]', 'ا', cleaned_text)
    cleaned_text = re.sub('ؤ', 'ء', cleaned_text)
    cleaned_text = re.sub('ئ', 'ء', cleaned_text)

    cleaned_text = re.sub(r'\u0610', '', cleaned_text)  # ARABIC SIGN SALLALLAHOU ALAYHE WASSALLAM
    cleaned_text = re.sub(r'\u0611', '', cleaned_text)  # ARABIC SIGN ALAYHE ASSALLAM
    cleaned_text = re.sub(r'\u0612', '', cleaned_text)  # ARABIC SIGN RAHMATULLAH ALAYHE
    cleaned_text = re.sub(r'\u0613', '', cleaned_text)  # ARABIC SIGN RADI ALLAHOU ANHU
    cleaned_text = re.sub(r'\u0614', '', cleaned_text)  # ARABIC SIGN TAKHALLUS

    cleaned_text = re.sub(r'\u06D6', '', cleaned_text)  # ARABIC SMALL HIGH LIGATURE SAD WITH LAM WITH ALEF MAKSURA
    cleaned_text = re.sub(r'\u06D7', '', cleaned_text)  # ARABIC SMALL HIGH LIGATURE QAF WITH LAM WITH ALEF MAKSURA
    cleaned_text = re.sub(r'\u06D8', '', cleaned_text)  # ARABIC SMALL HIGH MEEM INITIAL FORM
    cleaned_text = re.sub(r'\u06D9', '', cleaned_text)  # ARABIC SMALL HIGH LAM ALEF
    cleaned_text = re.sub(r'\u06DA', '', cleaned_text)  # ARABIC SMALL HIGH JEEM
    cleaned_text = re.sub(r'\u06DB', '', cleaned_text)  # ARABIC SMALL HIGH THREE DOTS
    cleaned_text = re.sub(r'\u06DC', '', cleaned_text)  # ARABIC SMALL HIGH SEEN

    cleaned_text = re.sub(r'\u06DD', '', cleaned_text)  # ARABIC END OF AYAH
    cleaned_text = re.sub(r'\u06DE', '', cleaned_text)  # ARABIC START OF RUB EL HIZB
    cleaned_text = re.sub(r'\u06DF', '', cleaned_text)  # ARABIC SMALL HIGH ROUNDED ZERO
    cleaned_text = re.sub(r'\u06E0', '', cleaned_text)  # ARABIC SMALL HIGH UPRIGHT RECTANGULAR ZERO
    cleaned_text = re.sub(r'\u06E1', '', cleaned_text)  # ARABIC SMALL HIGH DOTLESS HEAD OF KHAH
    cleaned_text = re.sub(r'\u06E2', '', cleaned_text)  # ARABIC SMALL HIGH MEEM ISOLATED FORM
    cleaned_text = re.sub(r'\u06E3', '', cleaned_text)  # ARABIC SMALL LOW SEEN
    cleaned_text = re.sub(r'\u06E4', '', cleaned_text)  # ARABIC SMALL HIGH MADDA
    cleaned_text = re.sub(r'\u06E5', '', cleaned_text)  # ARABIC SMALL WAW
    cleaned_text = re.sub(r'\u06E6', '', cleaned_text)  # ARABIC SMALL YEH
    cleaned_text = re.sub(r'\u06E7', '', cleaned_text)  # ARABIC SMALL HIGH YEH
    cleaned_text = re.sub(r'\u06E8', '', cleaned_text)  # ARABIC SMALL HIGH NOON
    cleaned_text = re.sub(r'\u06E9', '', cleaned_text)  # ARABIC PLACE OF SAJDAH
    cleaned_text = re.sub(r'\u06EA', '', cleaned_text)  # ARABIC EMPTY CENTRE LOW STOP
    cleaned_text = re.sub(r'\u06EB', '', cleaned_text)  # ARABIC EMPTY CENTRE HIGH STOP
    cleaned_text = re.sub(r'\u06EC', '', cleaned_text)  # ARABIC ROUNDED HIGH STOP WITH FILLED CENTRE
    cleaned_text = re.sub(r'\u06ED', '', cleaned_text)  # ARABIC SMALL LOW MEEM

    # Remove Arabic Shadda with Damma (often used incorrectly in place of regular Shadda in some texts)
    cleaned_text = re.sub(r'\uFE7C', '', cleaned_text)  # ARABIC SHADDA WITH DAMMA (Presentation Form)

    # Remove Arabic diacritics using their Unicode points
    cleaned_text = re.sub(r'\u0651', '', cleaned_text)  # ARABIC SHADDA
    cleaned_text = re.sub(r'\u064B', '', cleaned_text)  # ARABIC FATHATAN
    cleaned_text = re.sub(r'\u064C', '', cleaned_text)  # ARABIC DAMMATAN
    cleaned_text = re.sub(r'\u064D', '', cleaned_text)  # ARABIC KASRATAN
    cleaned_text = re.sub(r'\u0652', '', cleaned_text)  # ARABIC SUKUN
    cleaned_text = re.sub(r'\u08F0', '', cleaned_text)  # ARABIC OPEN FATHATAN
    cleaned_text = re.sub(r'\u08F1', '', cleaned_text)  # ARABIC OPEN DAMMATAN
    cleaned_text = re.sub(r'\u08F2', '', cleaned_text)  # ARABIC OPEN KASRATAN
    cleaned_text = re.sub(r'\u064E', '', cleaned_text)  # ARABIC FATHA
    cleaned_text = re.sub(r'\u064F', '', cleaned_text)  # ARABIC DAMMA
    cleaned_text = re.sub(r'\u0650', '', cleaned_text)  # ARABIC KASRA
    cleaned_text = re.sub(r'\uFDF2', '', cleaned_text)  # ARABIC LIGATURE ALLAH ISOLATED FORM
    cleaned_text = re.sub(r'\uFDFD', '', cleaned_text)  # ARABIC LIGATURE BISMILLAH AR-RAHMAN AR-RAHEEM
    cleaned_text = re.sub(r'\uFDFA', '', cleaned_text)  # ARABIC LIGATURE SALLALLAHOU ALAYHE WASALLAM
    cleaned_text = re.sub(r'\uFDFB', '', cleaned_text)  # ARABIC LIGATURE JALLAJALALOUHOU
    cleaned_text = re.sub(r'\uFDFC', '', cleaned_text)  # RIAL SIGN
    cleaned_text = re.sub(r'\uFDFF', '', cleaned_text)  # ARABIC LIGATURE AR-RAHMAN AR-RAHEEM

    cleaned_text = re.sub(r'\u0670', '', cleaned_text)

    # Remove non-textual elements
    cleaned_text = re.sub('[^\u0600-\u06FF]', ' ', cleaned_text)  # Keep Arabic characters only
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # Remove multiple spaces

    return cleaned_text

# Apply Normalization
Normalize_Clean_Text = clean_quran_df['Ayah Text'].apply(clean_normalize_quran_text)

# Apply Tokenization
clean_quran_df['Clean_Tokens'] = Normalize_Clean_Text.apply(tokenize)

# Drop not needed columns
#clean_quran_df = clean_quran_df.drop(['Clean Text'], axis=1)

# Convert Suran and Verse number to Integer Type.
clean_quran_df['Juz Num'] = clean_quran_df['Juz Num'].astype(int)
clean_quran_df['Sura Num'] = clean_quran_df['Sura Num'].astype(int)
clean_quran_df['Ayah Num'] = clean_quran_df['Ayah Num'].astype(int)

# Save the cleaned dataset to a new Excel file
# ==========================
cleaned_dataset_path = '/content/drive/MyDrive/Artificial Intelligence/Class: Natural Language Processing NLP/Project /Phase 1/Output/cleaned_QuranTxT.xlsx'
directory = os.path.dirname(cleaned_dataset_path)
if not os.path.exists(directory):
    os.makedirs(directory)
clean_quran_df.to_excel(cleaned_dataset_path, index=False)
# ===========================
```

# **Tokenization and Annotation**

```python
import pandas as pd
import spacy
from spacy.lang.ar import Arabic
from typing import Dict, List
from tqdm.auto import tqdm

# Initialize spaCy's Arabic model
nlp = Arabic()

def prefix_bismillah(sura, ayah, ayah_text):
    bismillah = "بِسْمِ اللَّهِ الرَّحْمَـٰنِ الرَّحِيمِ"
    # The Bismillah phrase should only be removed if it's at the start of the first ayah in suras other than the first
    if sura != 1 and ayah == 1:
        if ayah_text.startswith(bismillah):
            return ayah_text[len(bismillah):].strip()
    return ayah_text

def map_annotations_to_verse_text(juz, sura, ayah, sura_name, QuranCorpus, verse_text):
    adjusted_verse_text = prefix_bismillah(sura, ayah, verse_text)
    token_tag_tuples = []
    word_tags = {}

    # Tokenize the adjusted verse text
    doc = nlp(adjusted_verse_text)
    tokens = [token.text for token in doc]

    # Temporary storage to collect tags for each word number
    word_components = {}

    for entry in QuranCorpus:
        try:
            location_parts = entry['location'].strip("()").split(":")
            if len(location_parts) < 4 or not all(part.isdigit() for part in location_parts):
                continue  # Skip entries with invalid or incomplete location data
            loc_sura, loc_ayah, word_number, part_number = map(int, location_parts)
            if loc_sura == sura and loc_ayah == ayah:
                # Collecting tags for each word number
                word_components.setdefault(word_number, []).append(entry['tag'])
        except KeyError:
            continue  # Skip entries with missing keys

    # Iterate over the tokens to match them with word components
    for token_index, token in enumerate(tokens, start=1):  # Start indexing from 1
        tags = word_components.get(token_index, [])
        tag_str = '+'.join(tags) if tags else 'None'

        # Construct the location string including the token index and Sura name
        token_location = f"T{token_index}A{ayah}S{sura}J{juz}"
        token_sura = f"{sura_name}"
        token_tag_tuples.append((token, tag_str, token_location, token_sura, ))
        word_tags[token] = tag_str


    return token_tag_tuples, word_tags

def process_quran_text(quran_df, QuranCorpus):
    # Initialize the new columns if they don't exist
    if 'Token_POS' not in quran_df.columns:
        quran_df['Token_POS'] = None
    if 'POS_Tags' not in quran_df.columns:
        quran_df['POS_Tags'] = None
    if 'Token_POS_Location' not in quran_df.columns:
        quran_df['Token_POS_Location'] = None

    for index, row in tqdm(quran_df.iterrows(), total=len(quran_df), desc="Processing Quran Text"):
        ayah_text = row['Clean Text']
        juz = row['Juz Num']
        sura = row['Sura Num']
        ayah = row['Ayah Num']
        sura_name = row['Sura Nam']

        # Map POS tags to verse text using the adjusted function
        _, word_tags = map_annotations_to_verse_text(juz, sura, ayah, sura_name, QuranCorpus, ayah_text)
        word_tags_list, _ = map_annotations_to_verse_text(juz, sura, ayah, sura_name, QuranCorpus, ayah_text)
        quran_df.at[index, 'Token_POS'] = str(word_tags)
        quran_df.at[index, 'POS_Tags'] = list(word_tags.values())
        quran_df.at[index, 'Token_POS_Location'] = str(word_tags_list)

# Assuming `CleanQuranTxT` and `QuranCorpus` are already defined and loaded
QURAN_DF = pd.DataFrame(clean_quran_df)
process_quran_text(QURAN_DF, QURAN_CORPUS)
```

# **Tokens Mapping**

Mapping each token to its POS Tag, Sura Number, and Ayah Number

```python
import pandas as pd
import ast

# Initialize an empty list to hold dictionaries for the new DataFrame
verse_tags_list = []

# Iterate over the rows of the original DataFrame
for word_tags_str in QURAN_DF['Token_POS_Location']:
    try:
        # Safely convert the string representation to a list of tuples
        word_tags_tuples = ast.literal_eval(word_tags_str)

        # Now, iterate over each tuple in the list
        for token, pos_tag, location, sura_name,   in word_tags_tuples:
            # Create a new row with the token, POS tag, Sura, and Ayah and add it to the list
            verse_tags_list.append({'Token': token,'POS Tag': pos_tag, 'Location': location, 'Sura Name': sura_name})

    except (ValueError, SyntaxError) as e:
        # Handle cases where conversion fails
        print(f"Error converting string to list of tuples: {e}")

_df = pd.DataFrame(verse_tags_list)

# Save the cleaned dataset to a new Excel file
# ==========================
updated_df_path = '/content/drive/MyDrive/Artificial Intelligence/Class: Natural Language Processing NLP/Project /Phase 1/Output/_df.xlsx'
directory = os.path.dirname(updated_df_path)
if not os.path.exists(directory):
    os.makedirs(directory)
_df.to_excel(updated_df_path, index=False)
# ===========================
```
