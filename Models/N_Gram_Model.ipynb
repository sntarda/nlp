```python
import pandas as pd
import numpy as np
from collections import defaultdict
import random

# Load the dataset
def load_data(file_path):
    return pd.read_excel(file_path)

def parse_location(location):
    try:
        token_part, ayah_part, sura_part = location.split('A')[0], location.split('A')[1].split('S')[0], location.split('S')[1].split('J')[0]
        ayah = int(ayah_part)
        sura = int(sura_part)
        return sura, ayah
    except (IndexError, ValueError) as e:
        print(f"Error parsing location: {location} - {e}")
        return None, None

def create_sura_id_map(df):
    sura_names = df['Sura Name'].unique()
    num_to_sura_name = {}
    for name in sura_names:
        sample_location = df[df['Sura Name'] == name]['Location'].iloc[0]
        sura_number, _ = parse_location(sample_location)
        num_to_sura_name[sura_number] = name
        #print(f"Location: {sample_location}, Extracted Sura: {sura_number}, Name: {name}")  # Debugging line
    return num_to_sura_name

def prepare_ngram_data(df, n=3):
    data = []
    tokens = df['Token'].tolist()
    locations = df['Location'].tolist()
    for i in range(len(tokens) - n):
        sequence = tokens[i:i+n]
        next_word = tokens[i+n]
        next_location = locations[i+n]
        data.append((sequence, next_word, next_location))
    return data

def train_ngram_model(data):
    model = defaultdict(lambda: [])
    for sequence, next_word, location in data:
        model[tuple(sequence)].append((next_word, location))
    return model

def predict_next_sequence(model, input_sequence, sequence_length=3):
    current_input = input_sequence[-3:]  # use the last three words
    predicted_sequence = []
    sura, ayah = None, None
    for _ in range(sequence_length):
        current_tuple = tuple(current_input[-3:])
        if current_tuple in model:
            predictions = model[current_tuple]
            chosen_prediction = random.choice(predictions)
            next_word, location = chosen_prediction
            sura, ayah = parse_location(location)
            predicted_sequence.append(next_word)
            current_input.append(next_word)
        else:
            predicted_sequence.append("You probably made a spelling error")
            break
    return predicted_sequence, sura, ayah

def interactive_command_line_app(model, num_to_sura_name):
    print("Welcome to the Quranic Text Prediction Tool!")
    print("Type a sequence of three or more words from the Quran and the model will predict the next words.")
    print("Type 'exit' to quit the application.\n")

    while True:
        initial_text = input("Enter your text: ")
        if initial_text.lower() == 'exit':
            print("Exiting the application. Goodbye!")
            break

        input_sequence = initial_text.split()
        if len(input_sequence) < 4:
            print("Error: Input must be at least three words long.\n")
            continue

        predicted_sequence, sura, ayah = predict_next_sequence(model, input_sequence, 3)
        output_text = ' '.join(predicted_sequence)

        if sura and ayah:
            sura_name = num_to_sura_name.get(sura, f"Unknown Sura for ID {sura}")
            print(f"Predicted continuation: {output_text}")
            print(f"This corresponds to Sura: {sura_name} (Sura number {sura}), Ayah {ayah}\n")
        else:
            print(output_text + "\n")


if __name__ == "__main__":
    quran_df = load_data('/content/drive/MyDrive/Quran_df.xlsx')

    # Create sura to ID map and extract data for n-gram model training
    num_to_sura_name = create_sura_id_map(quran_df)  # This function should now work correctly
    ngram_data = prepare_ngram_data(quran_df, n=4)
    ngram_model = train_ngram_model(ngram_data)

    # Start the interactive command line application with the correct Sura names
    interactive_command_line_app(ngram_model, num_to_sura_name)
```
